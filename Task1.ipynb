```python
# Extracting required columns 
countries = data['Country Name']
population_1996 = data['1996']

# List of countries 
selected_countries = ['China', 'India', 'Russia', 'Japan', 'Canada', 'Australia', 'Germany', 'France', 'Samoa', 'South Korea']

# Filtering the DataFrame for the selected countries
selected_data = data[data['Country Name'].isin(selected_countries)]

# Sorting
sorted_data = selected_data.sort_values(by='1996', ascending=False)

# Creating a bar chart
plt.figure(figsize=(11, 6))
plt.bar(sorted_data['Country Name'], sorted_data['1996'], color='pink')
plt.xlabel('Country')
plt.ylabel('Population in 1996')
plt.title(f'Distribution of Population of selected countries in 1996')
plt.xticks(rotation=45)
plt.show()
```
